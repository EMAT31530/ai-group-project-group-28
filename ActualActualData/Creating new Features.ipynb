```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from imblearn.over_sampling import SMOTE, RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
```

```python
df = pd.read_csv('training_set.csv')
df.head()
```

```python
X, y = df.iloc[:,:-1], df.iloc[:, -1]
```

```python
#SMOTE
sm = SMOTE(random_state=0)
X_smote, y_smote = sm.fit_resample(X,y)

X_smote['default.payment.next.month'] = y_smote
```

```python
#RandomOverSample
ros = RandomOverSampler(random_state=0)
X_ros, y_ros = ros.fit_resample(X,y)

X_ros['default.payment.next.month'] = y_ros
```

```python
#RandomOverSample
rus = RandomUnderSampler(random_state=0)
X_rus, y_rus = rus.fit_resample(X,y)

X_rus['default.payment.next.month'] = y_rus
```

```python
print(df['default.payment.next.month'].value_counts())
print(X_smote['default.payment.next.month'].value_counts())
print(X_ros['default.payment.next.month'].value_counts())
print(X_rus['default.payment.next.month'].value_counts())
```

```python
X_ros.to_csv('training_ROS.csv', index=False)
X_rus.to_csv('training_RUS.csv', index=False)
X_smote.to_csv('training_SMOTE.csv', index=False)
```
