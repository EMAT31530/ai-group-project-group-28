```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from imblearn.over_sampling import SMOTE, RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler, CondensedNearestNeighbour
```

```python
df = pd.read_csv('training_set.csv')
df.head()
```

```python
X, y = df.iloc[:,:-1], df.iloc[:, -1]
```

```python
#SMOTE
sm = SMOTE(random_state=0)
X_smote, y_smote = sm.fit_resample(X,y)

X_smote['credit_card_default'] = y_smote
```

```python
#RandomOverSample
ros = RandomOverSampler(random_state=0)
X_ros, y_ros = ros.fit_resample(X,y)

X_ros['credit_card_default'] = y_ros
```

```python
#RandomOverSample
rus = RandomUnderSampler(random_state=0)
X_rus, y_rus = rus.fit_resample(X,y)

X_rus['credit_card_default'] = y_rus
```

```python
print(df['credit_card_default'].value_counts())
print(X_smote['credit_card_default'].value_counts())
print(X_ros['credit_card_default'].value_counts())
print(X_rus['credit_card_default'].value_counts())
```

```python
X_ros.to_csv('training_ROS.csv', index=False)
X_rus.to_csv('training_RUS.csv', index=False)
X_smote.to_csv('training_SMOTE.csv', index=False)
```

```python
#RandomOverSample
cnn = CondensedNearestNeighbour(random_state=0)
X_cnn, y_cnn = cnn.fit_resample(X,y)

X_cnn['default.payment.next.month'] = y_cnn
```

```python
X_cnn.to_csv('training_cnn.csv', index=False)
```
